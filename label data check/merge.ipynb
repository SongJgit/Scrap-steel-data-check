# 所有的提取出来放到images和labels里面

```python
path = 'data2'
target = '1'
```

```python
if not os.path.exists('images'):
    os.makedirs('images')
    os.makedirs('labels')
```

```python
def merge(path):
    rootfile_list = os.listdir(path)
    for i in rootfile_list:
        img_file_list = os.listdir(os.path.join(path,i,'images'))
        txt_file_list = os.listdir(os.path.join(path,i,'labels'))
        for txt_fi in txt_file_list:
            shutil.copy(os.path.join(path,i,'labels',txt_fi),os.path.join('labels'))
        for img_fi in img_file_list:
            shutil.copy(os.path.join(path,i,'images',img_fi),os.path.join('images'))
    pass
```

```python
merge(path)
```

# 删掉labels里的txt

```python
path = 'data'
```

```python
file_list = os.listdir(os.path.join('labels'))
```

```python
for fi in file_list:
    sep = os.path.splitext(fi)
    if sep[1] == ".txt":
        os.remove(os.path.join('labels',fi))
```

```python
def checkxml(path):
    #检查是否有错误的xml
    rootfile_list = os.listdir(path)
    for i in rootfile_list:
        img_file_list = os.listdir(os.path.join(path,i,'images'))
        txt_file_list = os.listdir(os.path.join(path,i,'labels'))
        for txt_fi in txt_file_list:
            sep = os.path.splitext(txt_fi)
            if sep[1] == ".txt":
                os.remove(os.path.join(path,i,'labels',txt_fi))
                print('{}已删除'.format(os.path.join(path,i,'labels',txt_fi)))
                
```

```python
checkxml('data')
```

```python
def checklength(path):
    #检查每个文件的长度是否一致
    rootfile_list = os.listdir(path)
    print('正数:img多于labels；负数:labels多于img')
    for i in rootfile_list:
        img_file_list = os.listdir(os.path.join(path,i,'images'))
        txt_file_list = os.listdir(os.path.join(path,i,'labels'))
        if len(img_file_list)==len(txt_file_list):
            pass
        else:
            print('{}images与labels数量不一致,缺失的数量为{}'.format(i,len(img_file_list)-len(txt_file_list)))
```
